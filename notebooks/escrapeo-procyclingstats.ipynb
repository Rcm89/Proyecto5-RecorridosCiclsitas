```python
import requests
from bs4 import BeautifulSoup
import pandas as pd



```

```python
import os
from tqdm import tqdm
```

```python
lista_carreras=["giro-d-italia", "tour-de-france", "vuelta-a-espana"]
```

```python
for carrera in lista_carreras:
    url = f"https://www.procyclingstats.com/race/{carrera}/route/climbs"

    response = requests.get(url)

    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table")

    headers = [th.text.strip() for th in table.find_all("th")]

    rows = []
    for row in table.find_all("tr")[1:]:  # Omitir el encabezado
        cells = [td.text.strip() for td in row.find_all("td")]
        rows.append(cells)
    
    if carrera == "giro-d-italia":
        df_puertos_giro = pd.DataFrame(rows, columns=headers)
    elif carrera == "tour-de-france":
        df_puertos_tour = pd.DataFrame(rows, columns=headers)
    elif carrera == "vuelta-a-espana":
        df_puertos_vuelta = pd.DataFrame(rows, columns=headers)


```

```python
for carrera in lista_carreras:
    ruta_datos = '../datos/datos_scrap_procyclingstats'
    nombre_archivo = f'df_puertos_{carrera}.csv'
    ruta_archivo = os.path.join(ruta_datos, nombre_archivo)
    if carrera == "giro-d-italia":
        df_puertos_giro.to_csv(ruta_archivo, index=False)
    elif carrera == "tour-de-france":
        df_puertos_tour.to_csv(ruta_archivo, index=False)
    elif carrera == "vuelta-a-espana":
        df_puertos_vuelta.to_csv(ruta_archivo, index=False)

```
